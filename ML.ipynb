```python
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
x_train=np.array([5,10,15,20,25,30])
y_train=np.array([15,25,35,45,55,65])
x_test=np.array([])
y_test=np.array([])

print("x_train :",x_train)
print("y_train :",y_train)
```

```python
# Generate  the random values 
import random
m = [random.randint(-50, 50) for i in range(1000)]
print(m)
c = [random.randint(-50, 50) for i in range(1000)]
print(c)
```

```python
plt.scatter(m, c,c='c')
plt.title("Plot")
plt.ylabel('m')
plt.xlabel('c')
```

```python
# Plot the data points
plt.scatter(x_train, y_train,c='c')
plt.title("Plot")
plt.ylabel('Y')
plt.xlabel('X')
```

```python
# define prediction
def compute_model_prediction(x, m, c):
    n = x.shape[0]
    f_wb = np.zeros(n)
    for i in range(n):
        f_wb[i] = m * x[i] + c
        
    return f_wb

# define cost funtion

def compute_cost(x, y, m, c): 
    n = x.shape[0] 
    
    cost_sum = 0 
    for i in range(n): 
        f_wb = m * x[i] + c   
        cost = (f_wb - y[i]) ** 2  
        cost_sum = cost_sum + cost  
    total_cost = (1 / (2 * n)) * cost_sum  

    return total_cost
```

```python
temp_m=[]
for i in m:
  temp_m.append(compute_model_prediction(x_train, i, 0))
```

```python
# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='c',label='Actual Values')
plt.scatter(x_train, y_train,c='b')
plt.title("Plot")
plt.ylabel('Y')
plt.xlabel('X')


# plot the prediction
for i in range(len(temp_m)):
  plt.plot(x_train, temp_m[i], c='r',label='Prediction')                     
                  
plt.show()                  
           
```

```python
# cost funtion for each m
temp_jmc=[]
for i in m:
  temp_jmc.append(compute_cost(x_train, y_train, i, 0))
print(temp_jmc)
```

```python
plt.plot(m,temp_jmc,c='b')
plt.title("cost funtion")
plt.ylabel('j(m)')
plt.xlabel('m')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(m, c, temp_jmc , c='r', marker='o')

# Adding labels to the x, y, and z axis
ax.set_xlabel('m')

ax.set_ylabel('c')
ax.set_zlabel('temp_jmc')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

plt.plot(m, c, 'r*')
plt.axis([0, 6, 0, 20])

for i, j,k in zip(m, c,temp_jmc):
  if k==min(temp_jmc):
    plt.text(i, j+0.5, '({}, {}, {})'.format(i, j,k))

plt.show()
```

```python
bOf_fmc=[]
for i,j,k in zip(m, c,temp_jmc):
  if k==min(temp_jmc):
    bOf_fmc.append(j)
print(bOf_fmc)
```

```python
#find cost

final_m=2
cost=[]
for c in bOf_fmc:
    cost.append(compute_cost(x_train, y_train, final_m, c))
print('cost:',cost)

#find c for minimum cost

final_c=bOf_fmc[cost.index(min(cost))]

print("C :",final_c)

```

```python
def model(x):
    y=final_m*x+final_c
    return y
```

```python
# Testing
y=model(35)
print('Y:',y)
```
